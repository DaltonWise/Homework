```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib in line
```

```python

data = pd.read_csv('/content/ORES table formatted.csv')

#This is assignining the info as a data frame so it can be parced properly.
df = pd.DataFrame(data)

#showing myself how the data looks here.
df.head()
```

**Dropping Columns**

```python
#I dropped each of these columns because they lead to outliers in the data and take away from some of the correlation in graphs.
#I do not think it is valuable to have the total of both, men, and women. These values are much larger and do not help show any trend.
df.drop(['Both_Total', 'Male_Total', 'Female_Total'], axis=1, inplace=True)
df.head()
```

```python
#These two commands remove the rows indicated.
#Below removes row 0.
df.drop(0, axis=0, inplace = True)
#Below removes row 2.
df.drop(2, axis=0, inplace = True)
df.head()
```

**Transforming data types**

```python
#This is showing the data types of each column in the data set.

#I am going to change some of them to integers to show that I know how.
df.dtypes
```

```python
#I chose to convert the columns representing masters degrees to convert from float to integer.


#It is actually better to use integers over floats when you can because it improve computational speed.
#Integers are stored as fixed values while floats can be stored in a number of ways that all require computation before being identified.
#https://discuss.codecademy.com/t/if-most-numbers-can-be-represented-as-a-float-value-why-would-we-need-to-use-integers/347282#:~:text=The%20advantage%20of%20int%20over,just%20in%20assessing%20the%20value.
#We also literally do not need them to be floats because none of them have decimal values.


#Both
df['B_Bachelor_degree'] = df['B_Bachelor_degree'].astype(int)
df['B_Master_degree'] = df['B_Master_degree'].astype(int)
df['B_Professional_degree'] = df['B_Professional_degree'].astype(int)
df['B_Doctoral_degree'] = df['B_Doctoral_degree'].astype(int)
#Male
df['M_Bachelor_degree'] = df['M_Bachelor_degree'].astype(int)
df['M_Master_degree'] = df['M_Master_degree'].astype(int)
df['M_Professional_degree'] = df['M_Professional_degree'].astype(int)
df['M_Doctoral_degree'] = df['M_Doctoral_degree'].astype(int)
#Female
df['F_Bachelor_degree'] = df['F_Bachelor_degree'].astype(int)
df['F_Master_degree'] = df['F_Master_degree'].astype(int)
df['F_Professional_degree'] = df['F_Professional_degree'].astype(int)
df['F_Doctoral_degree'] = df['F_Doctoral_degree'].astype(int)
```

```python
#I used the print function to show the type had changed.
#I did not want include each one as it seemed unnecessary.
#This is me showing that the data types have been correctly altered.
print(df['B_Master_degree'].dtypes)
print(df['M_Master_degree'].dtypes)
print(df['F_Master_degree'].dtypes)
```

**Renaming Columns**

```python
#This is me renaming the degrees to something  that looks better.
df.rename(columns = {'B_Bachelor_degree' : 'B_Bachelor', 'B_Master_degree' : 'B_Master', 'B_Professional_degree' : 'B_Professional', 'B_Doctoral_degree' : 'B_Doctoral',
                     'M_Bachelor_degree' : 'M_Bachelor', 'M_Master_degree' : 'M_Master', 'M_Professional_degree' : 'M_Professional', 'M_Doctoral_degree' : 'M_Doctoral',
                     'F_Bachelor_degree' : 'F_Bachelor', 'F_Master_degree' : 'F_Master', 'F_Professional_degree' : 'F_Professional', 'F_Doctoral_degree' : 'F_Doctoral',
                    } , inplace = True)
df.columns
```

```python
#This is me renaming the masters degrees to something  that looks better.
df.rename(columns = {'B_Master_degree' : 'B_Master', 'M_Master_degree' : 'M_Master', 'F_Master_degree' : 'F_Master'}, inplace = True)
df.columns
```

**Create calculated data field**

```python
#This will show all the males in each age frame with a bachelors degree or higher. Those with a college education.
Male_Degree = ['M_Bachelor', 'M_Master', 'M_Professional', 'M_Doctoral']
df['Male_Degree'] = df[Male_Degree].sum(axis=1, skipna=False)
df['Male_Degree']
```

```python
#All women of a certain age bracket with atleast a bachelor degree.
Female_Degree = ['F_Bachelor', 'F_Master', 'F_Professional', 'F_Doctoral']
df['Female_Degree'] = df[Female_Degree].sum(axis=1, skipna=False)
df['Female_Degree']
```

**Drop Duplicates**

```python
#I am pretty sure I have no duplicate values.
#If I did they are gone.
df.drop_duplicates()
```

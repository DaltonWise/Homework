```python
#This was my first merge. I realized I could merge on the first three columns in both dataframes.
#To do that I had to make sure they all had the same data types.

#burden.merge(mental_health, on = 'Entity')
```

```python
burden['Year'] = burden['Year'].values.astype(str)
mental_health['Year'] = mental_health['Year'].values.astype(str)


#In this cell I am converting each of the year columns to a string. The reason I do this is because when using 'pd.merge' an error was raised stating I was trying to merge a string and a int64.
#I assume that in one of the two data frames the year was a string and the other it was an integer.

#https://sparkbyexamples.com/pandas/pandas-convert-columns-to-string-type/
```

```python
health_merged = pd.merge(burden, mental_health, on = ['Entity', 'Code', 'Year'])
```

```python
health_merged
```

```python
health_merged['Year'].unique()
#Trying to see the range of years in this data set. I would like to try to find world population data from that year range.
```

```python
population = world_population2.drop(['Indicator Name', 'Indicator Code'], axis = 1)

#Dropping those two columns so that I can hopefully unpivot this data.
```

```python
population.head()
```

```python
population = population.set_index(["Country Name", "Country Code"])
population.head()

#Because I am trying to pivot the data, I am converting the country name and country code to the index.
#This will allow me to make the year one single column.
```

```python
population2 = population.stack().reset_index()
```

```python
population2
```

```python
population2.columns = ['Entity', 'Code', 'Year', 'Total_Population']
#Changing the names to match the previous two data sets.
```

```python
population2.head()
```

```python
second_merge = pd.merge(health_merged, population2, how = 'left', on = ['Entity', 'Code', 'Year'])

#I used a left merge so that I would only be pulling records from 'population2' that matched records from 'health_merged'.
#I can see by comparing that 'second_merge' and 'health_merged' both have 23,260 rows. I was going to try to right an expression to show that.
#The only thing I could think to do was try to use the shape function to prove they were both the same but they actually have a different number of columns so I cant.
#Either way my point stands that I only returned the records matching from health_merged.
```
